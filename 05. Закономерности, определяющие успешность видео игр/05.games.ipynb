### Описание проекта

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

### 2. Подготовка данных

Изменим название столбцов

```python
data.columns # перечень названий столбцов таблицы 
```

```python
data.columns = data.columns.str.lower()
```

```python
data.columns
```

```python
data.duplicated().sum() # подсчёт явных дубликатов
```

```python
data[data['user_score'] == 'tbd']
```

Некоторые значения столбца user_score имеют строковые значения tbd.
'To Be Determined' - 'Будет определено'. Т.е. это то же самое, что неизвестное значение. Поэтому, можно заменить значение 'tbd' на 'NaN' и изменить тип данных

```python
data['user_score'] = data['user_score'].replace('tbd', np.nan, regex=True)
data['user_score'] = data['user_score'].astype('float64')
```

В year_of_release заменим тип данных на int

```python
data['year_of_release'] = data['year_of_release'].astype('Int64')
```

Удалим пустые значения в столбцах name, genre(их всего 4)

```python
data = data.dropna(subset=['name', 'genre']).reset_index(drop=True)
```

Добавим отдельный столбец с суммарными продажами во всех регионах 

```python
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
```

#### 3.7 Соотношение выводов с продажами игр на других платформах

```python
def other_platform_matrix(name_of_platform):
    platform = actual_data[actual_data['platform'] == name_of_platform]
    fig, ax = plt.subplots(1, 2, figsize = (15,5))
    sns.scatterplot(x='user_score', y='total_sales', data=platform, ax=ax[0])
    sns.scatterplot(x='critic_score', y='total_sales', data=platform, ax=ax[1])
    fig.suptitle(name_of_platform, fontsize=15)
    ax[0].set(xlabel='Оценка пользователей')
    ax[1].set(xlabel='Оценка критиков')
    ax[0].set(ylabel='Количество продаж')
    ax[1].set(ylabel='Количество продаж')
    plt.show()
    
    users_correl = platform['user_score'].corr(platform['total_sales'])
    print('Корреляция между оценками пользователей и продажами', users_correl) 
    critic_correl = platform['critic_score'].corr(platform['total_sales'])
    print('Корреляция между оценками критиков и продажами', critic_correl)
```

```python
modern_platforms = ['PS4', 'XOne', '3DS', 'WiiU', 'PC']
for platforms in modern_platforms:
    other_platform_matrix(platforms)
```

Вывод:
Во всех случаях оценки критиков имеют влияние на количество продаж игр

### 4. Портрет пользователя каждого региона

#### 4.1 Самые популярные платформы (топ-5)

```python
games = actual_data.query('year_of_release >= 2015')
```

```python
top_regions_dict = {}
regions = ['na_sales', 'eu_sales', 'jp_sales']
for region in regions: 
    top_regions_dict[region] = (
        games.groupby('platform')[region].sum()
        .sort_values(ascending=False).head(5)
)
top_regions_dict
```

Для наглядности постороим графики:

```python
for i in regions:
    labels = top_regions_dict[i].index
    explode = [0.02, 0.05, 0.08, 0.08, 0.05]
    plt.figure(figsize=(12,6))
    plt.pie(x=top_regions_dict[i], labels = labels, autopct='%1.1f%%', shadow=True, explode=explode)
    plt.title(f'Продажи в {i}')
    plt.show()
```

- Самые популярные платформы в NA - PS4, XOne, 3DS, WiiU, X360
- Самые популярные платформы в EU - PS4, XOne, PC, 3DS, PS3 
- Самые популярные платформы в JP - 3DS, PS4, PSV, PS3, WiiU

Самой популярной платформой в Европе и Северной Америке является PS4. Скорее всего данная популярность связана с наличием эксклюзтвных видеоигр для данной платформы. В Японии же популярны игровые платформы собственного производства

#### 4.2 Самые популярные жанры (топ-5)

```python
top_genre_dict = {}
regions = ['na_sales', 'eu_sales', 'jp_sales']
for region in regions: 
    top_genre_dict[region] = (
        games.groupby('genre')[region].sum()
        .sort_values(ascending=False).head(5)
)
top_genre_dict
```

Для наглядности постороим графики:

```python
for i in regions:
    labels = top_genre_dict[i].index
    explode = [0.02, 0.05, 0.08, 0.08, 0.05]
    plt.figure(figsize=(12,6))
    plt.pie(x=top_regions_dict[i], labels = labels, autopct='%1.1f%%', shadow=True, explode=explode)
    plt.title(f'Продажи в {i}')
    plt.show()
```

- Самые популярные жанры в NA - Shooter, Action, Sport, Role-Playing, Misc
- Самые популярные жанры в EU - Shooter, Action, Sport, Role-Playing, Racing

### 5. Проверка гипотез

#### 5.1 Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Нулевая гипотеза H0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Альтернативная гипотеза H₁: Средние пользовательские рейтинги платформ Xbox One и PC различаются

```python
sample_1 = games[games['platform'] == 'XOne']['user_score'].dropna() # Оценки пользователей Xbox One
sample_2 = games[games['platform'] == 'PC']['user_score'].dropna() # Оценки пользователей PC
```

```python
print('Дисперсия XOne =', np.var(sample_1))
print('Дисперсия PC =', np.var(sample_2))     
```

Будем использовать метод ttest_ind, так как нам нужно проверить равенстве средних двух генеральных совокупностей. Дисперсии у генеральных совокупностей разные, поэтому параметр equal_var = False

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(sample_1, sample_2, equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вывод:
Так как принятая нулевая гипотеза не отвергается по результатам стат. теста, мы можем говорить, что разница между средними, если она есть, получена случайно. Между средними пользовательскими рейтингами платформ Xbox One и PC нет статистически значимой разницы. 

#### 5.2 Средние пользовательские рейтинги жанров Action и Sports разные.

Нулевая гипотеза H0: Средние пользовательские рейтинги жанров Action и Sports одинаковые
    
Альтернативная гипотеза H1: Средние пользовательские рейтинги жанров Action и Sports различаются.

```python
sample_3 = games[games['genre'] == 'Action']['user_score'].dropna() # пользовательские рейтинги жанров Action
sample_4 = games[games['genre'] == 'Sports']['user_score'].dropna() # пользовательские рейтинги жанров Sports
```

```python
print('Дисперсия Action =', np.var(sample_3))
print('Дисперсия Sports =', np.var(sample_4))     
```

Будем использовать метод ttest_ind, так как нам нужно проверить равенстве средних двух генеральных совокупностей. Дисперсии у генеральных совокупностей разные, поэтому параметр equal_var = False

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(sample_3, sample_4, equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вывод:
Средние пользовательские рейтингии жанров Action и Sports различаются статистически значимо.
